```
import numpy as np 
import matplotlib.pyplot as plt
```

```
x = np.arange(-5.0, 5.0, 0.1) #그림을 -5부터 5까지 보고 간격은 0.1이다.
```

```
x.shape # 데이터가 100개 -5에서 5까지 0.1씩이면 데이터 100개다
```

**Step function (계단함수)**

```
 def step_function(x):
    return np.array(x>0, dtype=np.int) # x값이 0보다 크면 참, 작으면 거짓
```

```
y = step_function(x) # x에 해당되는 데이터값이 y로 들어온다
```

```
y.shape # 넣은 x값이랑 똑같이 나와야함
```

```
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # y값 보여지는 범위 지정
plt.show()
```

**Sigmoid**

$f(x) = \cfrac{1}{1+e^{-x}}$

fraction은 분수를 의미

e=np.exp

```
def sigmoid1(x):
    return 1 / (1 +np.exp(-x))
```

```
x = np.arange(-5.0, 5.0, 0.1) # -5부터 5까지 간격은 0.1만큼
y = sigmoid1(x)
```

```
plt.plot(x, y)
plt.grid()
plt.show()
```

$f(x) = \cfrac{1}{1+e^{-c_{1}x}}$

```
def sigmoid2(c1, x):
    return 1 / (1 +np.exp(-c1*x)) 
```

```
y2 = sigmoid2(0.1, x) # 만약 c1을 1이라고 하면 차이가 없다. c1이 0.1일 때 거의 직선 형태
```

```
y2.shape
```

```
plt.plot(x,y2)
plt.grid()
```

```
y3 = sigmoid2(0.5, x) # c1이 0.5일 때 조금약한 S자 곡선형태
```

```
plt.plot(x,y3)
plt.grid()
```

```
y4 = sigmoid2(2, x) # c1이 2일때 S자곡선형태의 그래프
```

```
plt.plot(x,y4)
plt.grid()
```

```
for i in range(5): # c1값에 따라 달라지는 5개의 그래프를 표현
  c1=0.2*(i+1) # i가 0일때 그냥 0을 곱하는 거나 마찬가지이므로 1을 더해준다
  y2=sigmoid2(c1,x)
  plt.plot(x, y2)
  print(c1) # c1값의 추이
```

$f(x) = \cfrac{1}{1+e^{-c_{1}(x-c_{2})}}$

```
def sigmoid3(x, c1=1, c2=0): # 에러 방지 위해 값을 정의
    return 1 / (1 +np.exp(-c1*(x-c2))) 
```

```
y3=sigmoid3(x, 5, 0) # 만약 에러가 난다면 값을 모두 안줬기 때문. 참고로 값을 지정할 때 정의함수에서 지정한 대로 값을 지정해줘야 한다.
```

```
plt.plot(x,y3)
plt.grid()
```

```
for i in range(5): # c1값에 따라 달라지는 5개의 그래프를 표현
  c1=2 # c1은 어떻게 보면 기울기를 의미한다.
  c2=0.2*(i+1) # c2는 x축방향으로 평행이동 
  y2=sigmoid3(x, c1, c2)
  plt.plot(x, y2) # 즉 동일한 기울기(동일한 S자곡선형태)에서 x축방향으로 평행이동하는 형태의 그래프가 그려질 것이다.
  print(c1) # c1값의 추이
```
